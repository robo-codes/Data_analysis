### Data import and export into Python using a URL

```python
# save url in variable
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

```python
df1 = pd.read_csv(url, header = None)
```

```python
headers =["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels", "engine-location", 
         "wheel-base", "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size", "fuel-system",
         "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
```

```python
df1.columns = headers
```

```python
#Drop missing values on Price
df1.dropna(subset=["price"], axis = 0, inplace =True)

```

```python
#Replace ? with NaN
df1.replace("?", np.nan, inplace = True)
```

```python
df1.head()
```

```python
# Check for missing values
missing_data = df1.isnull()
missing_data.head()
```

```python
# Count missing values
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

#### Question: Which columns have the missing values? 

```python
#Replace missing values by first calculating the mean and then replace for normalized losses
mean_loss = df1["normalized-losses"].astype("float").mean()
df1["normalized-losses"].replace(np.nan, mean_loss, inplace=True)
```

```python
df1["normalized-losses"].head()
```

```python
# Do the same for "stroke", "bore", "horsepower", "peak-rpm" columns
mean_stroke = df1["stroke"].astype("float").mean()
df1["stroke"].replace(np.nan, mean_stroke, inplace=True)

mean_bore = df1["bore"].astype("float").mean()
df1["bore"].replace(np.nan, mean_bore, inplace=True)

mean_power = df1["horsepower"].astype("float").mean()
df1["horsepower"].replace(np.nan, mean_power, inplace=True)

mean_rpm = df1["peak-rpm"].astype("float").mean()
df1["peak-rpm"].replace(np.nan, mean_rpm, inplace=True)
```

```python
df1.head()
```

```python
#To see which values are present in a particular column
df1['num-of-doors'].value_counts()
```

```python
# Replace with four doors
df1["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#drop all rows that do not have price data
df1.dropna(subset=["price"], axis=0, inplace=True)
```

```python
df1.head()
```

```python
df1
```

```python
#reset index 
df1.reset_index(drop=True, inplace=True)
```

```python
df1
```

```python
#Convert data types to proper format
df1[["bore", "stroke", "peak-rpm", "price"]] = df1[["bore", "stroke", "peak-rpm", "price"]].astype("float")
df1[["normalized-losses"]] = df1[["normalized-losses"]].astype("int")

```

```python
df1.dtypes
```

```python
#Normalize length, height and width columns
# replace (original value) by (original value)/(maximum value)
df1['length'] = df1['length']/df1['length'].max()
df1['width'] = df1['width']/df1['width'].max()
df1['height']= df1['height']/df1['height'].max()
```

```python
#Lets do binning of horsepower variable into 3 bins
#First convert it into the correct data type

df1[["horsepower"]] = df1[["horsepower"]].astype("int")
```

```python
# Import libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.distplot(df1["horsepower"], bins=3)
```

```python
#Rename columns
dv1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dv1.head()
```

```python
df1.columns
```

```python
# merge data frame "df" and "dummy_variable_1" 
df1 = pd.concat([df1, dv1], axis=1)
```

```python
# drop the original fuel type column
df1.drop("fuel-type", axis = 1, inplace=True)
```

```python
df1.columns
```

```python
df1['aspiration'].value_counts()
```

#### Do the same for aspiration column std = 0 and turbo =1

```python
# get indicator variables of aspiration and assign it to data frame "dummy_variable_2"
dv2 = pd.get_dummies(df1['aspiration'])

# change column names for clarity
dv2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

# show first 5 instances of data frame "dummy_variable_1"
dv2.head()
```

```python
#merge the new dataframe to the original datafram
df1 = pd.concat([df1, dv2], axis=1)

# drop original column "aspiration" from "df"
df1.drop('aspiration', axis = 1, inplace=True)
```

### Basic Graph


```python
#A default simple plot
x = np.array([1, 2, 3])
y = x * 2
plt.plot(x, y)        # plot x and y using default line style and color
```

```python
#A default simple plot
x = np.array([1, 2, 3])
y = x * 2
plt.plot(y, 'r+')     # plot y with red plusses
```

```python
#A default simple plot
x = np.array([1, 2, 3])
y = x * 2
plt.plot(y, 'g+')     # plot y with red plusses
```

```python
#creating a more readable and better plot
x = np.array([1 ,2 , 3])
y = x * 2

plt.plot(x, y, label = 'legend 2x')
plt.title('Title of the graph!', color = 'red')    
plt.xlabel('Label of the X Axes', color = "orange")
plt.ylabel('Lable of the Y Axes',color= "magenta")
plt.legend()
plt.show()
```

**Customization**
Style, Font, Size, legend, Color, Labels, Ticks

```python
#Customising the plot
x = np.array([1 ,2 , 3])
y = x * 2

#print(plt.style.available)
#plt.style.use('fivethirtyeight')

plt.plot(x, y, label = 'legend 2x', color='green', marker='o', linestyle='dashed',linewidth=2, markersize=8)
#plt.plot(x, y ,'go--', label = 'legend 2x',linewidth=2, markersize=8)     # 'go--' uses format of shorthand annotation of (color, marker, line) needs to be in order

plt.title('Title of the graph!',fontdict={'fontname': 'Arial', 'fontsize': 20}, color = "red")     #fontdict is a dictionary controlling the appearance of the title
plt.xlabel('Label of the X Axes', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20}, color = "orange")
plt.ylabel('Lable of the Y Axes',fontsize = 20, fontname= 'Arial Black', color= "magenta")

#plt.xticks(x)            #set the current tick locations and labels of the x-axis. 
plt.xticks(x , ['One', 'Two', 'Three'],rotation=20)  # Set text labels and properties.
#plt.tick_params(axis='x', rotation=90, colors='red')
plt.yticks(y)             #set the current tick locations and labels of the y-axis. 

plt.legend()
plt.show()
```
